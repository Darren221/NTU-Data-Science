```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('stores.csv',encoding = 'utf-8')
df = df[df.分公司狀態 == 1].drop_duplicates(subset=['分公司地址']).reset_index(drop=True)

display(df.head())
df.describe()
```

```python
#全台便利商店
df['縣市'] = df['分公司地址'].str[:3]
df['縣市'] = df['縣市'].str.replace('巿','市')
for i in range(df.shape[0]):
    if df.loc[i, '縣市'] == '台北市':
        df.loc[i, '縣市'] = '臺北市'
    if df.loc[i, '縣市'] == '臺北縣' or df.loc[i, '縣市'] == '台北縣':
        df.loc[i, '縣市'] = '新北市'
    if df.loc[i, '縣市'] == '桃園縣':
        df.loc[i, '縣市'] = '桃園市'
    if df.loc[i, '縣市'] == '臺中縣':
        df.loc[i, '縣市'] = '臺中市'
    if df.loc[i, '縣市'] == '臺南縣':
        df.loc[i, '縣市'] = '臺南市'
    if df.loc[i, '縣市'] == '高雄縣':
        df.loc[i, '縣市'] = '高雄市'
    if df.loc[i, '縣市'] == '苗栗市':
        df.loc[i, '縣市'] = '苗栗縣'
    if df.loc[i, '縣市'] == '彰化市':
        df.loc[i, '縣市'] = '彰化縣'
#display(df['縣市'])
total = df['縣市'].value_counts().rename_axis('縣市').reset_index(name = '便利商店')
#display(total)
print(df['公司名稱'].value_counts())
```

```python
seven = df[df.公司名稱 == '統一超商股份有限公司'].reset_index(drop=True)
total = total.merge(seven['縣市'].value_counts().rename_axis('縣市').reset_index(name = '7-11') ,how = 'left' ,left_on = '縣市' ,right_on = '縣市')
total['7-11'] = total['7-11'] / total['便利商店']

OK = df[df.公司名稱 == '來來超商股份有限公司'].reset_index(drop=True)
total = total.merge(OK['縣市'].value_counts().rename_axis('縣市').reset_index(name = 'OK') ,how = 'left' ,left_on = '縣市' ,right_on = '縣市').fillna(0)
total['OK'] = total['OK'] / total['便利商店']

family = df[df.公司名稱 == '全家便利商店股份有限公司'].reset_index(drop=True)
total = total.merge(family['縣市'].value_counts().rename_axis('縣市').reset_index(name = 'Family') ,how = 'left' ,left_on = '縣市' ,right_on = '縣市').fillna(0)
total['Family'] = total['Family'] / total['便利商店']

hi = df[df.公司名稱 == '萊爾富國際股份有限公司'].reset_index(drop=True)
total = total.merge(hi['縣市'].value_counts().rename_axis('縣市').reset_index(name = 'Hi-Life') ,how = 'left' ,left_on = '縣市' ,right_on = '縣市').fillna(0)
total['Hi-Life'] = total['Hi-Life'] / total['便利商店']

#display(total)

crime = pd.read_csv('crime_rates.csv')
#print(criminal)

#total = criminal.merge(total ,how = 'left' ,left_on = '縣市' ,right_on = '縣市')
total['刑案發生率'] = crime['刑案發生率']
total['竊盜案發生率'] = crime['竊盜案發生率']
total['暴力犯罪發生率'] = crime['暴力犯罪發生率']
display(total)

cooffense = total['刑案發生率'].corr(total['便利商店'])
print('Correlation between offense rate and proportion of convenience store = ', cooffense)
cooffense_711 = total['刑案發生率'].corr(total['7-11'])
print('Correlation between offense rate and proportion of 7-11 = ', cooffense_711)
cooffense_OK = total['刑案發生率'].corr(total['OK'])
print('Correlation between offense rate and proportion of OK = ', cooffense_OK)
cooffense_f = total['刑案發生率'].corr(total['Family'])
print('Correlation between offense rate and proportion of Family Mart = ', cooffense_f)
cooffense_hi = total['刑案發生率'].corr(total['Hi-Life'])
print('Correlation between offense rate and proportion of Hi-Life = ', cooffense_hi)
print()
colarceny = total['竊盜案發生率'].corr(total['便利商店'])
print('Correlation between larceny rate and proportion of convenience store = ', colarceny)
colarceny_711 = total['竊盜案發生率'].corr(total['7-11'])
print('Correlation between larceny rate and proportion of 7-11 = ', colarceny_711)
colarceny_OK = total['竊盜案發生率'].corr(total['OK'])
print('Correlation between larceny rate and proportion of OK = ', colarceny_OK)
colarceny_f = total['竊盜案發生率'].corr(total['Family'])
print('Correlation between larceny rate and proportion of Family Mart = ', colarceny_f)
colarceny_hi = total['竊盜案發生率'].corr(total['Hi-Life'])
print('Correlation between larceny rate and proportion of Hi-Life = ', colarceny_hi)
print()
coviolent = total['暴力犯罪發生率'].corr(total['便利商店'])
print('Correlation between violent crime rate and proportion of convenience store = ', coviolent)
coviolent_711 = total['暴力犯罪發生率'].corr(total['7-11'])
print('Correlation between violent crime rate and proportion of 7-11 = ', coviolent_711)
coviolent_OK = total['暴力犯罪發生率'].corr(total['OK'])
print('Correlation between violent crime rate and proportion of OK = ', coviolent_OK)
coviolent_f = total['暴力犯罪發生率'].corr(total['Family'])
print('Correlation between violent crime rate and proportion of Family Mart = ', coviolent_f)
coviolent_hi = total['暴力犯罪發生率'].corr(total['Hi-Life'])
print('Correlation between violent crime rate and proportion of Hi-Life = ', coviolent_hi)
```
