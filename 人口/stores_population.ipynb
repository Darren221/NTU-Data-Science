```python
# clean data of stores
stores_alive = stores_utf8[stores_utf8['分公司狀態']==1].copy().reset_index(drop=True)
stores_alive["行政區"] = stores_alive["分公司地址"].str.slice(0, 3)
stores_alive["行政區"] = stores_alive["行政區"].str.replace("巿", "市", regex=False)
stores_alive["行政區"] = stores_alive["行政區"].str.replace("台", "臺", regex=False)
stores_alive["行政區"] = stores_alive["行政區"].str.replace("苗栗市", "苗栗縣", regex=False)
stores_alive["行政區"] = stores_alive["行政區"].str.replace("彰化市", "彰化縣", regex=False)
stores_alive["行政區"] = stores_alive["行政區"].str.replace("臺北縣", "新北市", regex=False)
stores_alive["行政區"] = stores_alive["行政區"].str.replace("臺中縣", "臺中市", regex=False)
stores_alive["行政區"] = stores_alive["行政區"].str.replace("臺南縣", "臺南市", regex=False)
stores_alive["行政區"] = stores_alive["行政區"].str.replace("高雄縣", "高雄市", regex=False)
stores_alive["行政區"] = stores_alive["行政區"].str.replace("桃園縣", "桃園市", regex=False)
stores_alive = stores_alive.drop_duplicates(subset=["分公司地址"], keep="first")

# region
region = {"基隆市":"北", "新北市":"北", "臺北市":"北", "桃園市":"北", "新竹縣":"北", "新竹市":"北", "宜蘭縣":"北",
         "苗栗縣":"中", "臺中市":"中", "彰化縣":"中", "南投縣":"中", "雲林縣":"中",
         "嘉義縣":"南", "嘉義市":"南", "臺南市":"南", "高雄市":"南", "屏東縣":"南",
         "花蓮縣":"東", "臺東縣":"東", "澎湖縣":"外", "金門縣":"外", "連江縣":"外"}
stores_alive["區域"] = stores_alive["行政區"].map(region)

# brand
brand = {"統一超商股份有限公司":"7-11", "全家便利商店股份有限公司":"全家", "萊爾富國際股份有限公司":"萊爾富", "來來超商股份有限公司":"OK"}
stores_alive["品牌"] = stores_alive["公司名稱"].map(brand)

# sort
stores_alive["區域"] = pandas.Categorical(stores_alive["區域"], ["北", "中", "南", "東", "外"])

# save
#stores_alive.to_csv("stores_clean.csv")

test = stores_alive.groupby(["區域","行政區","品牌"]).size().unstack(level=-1)
display(test)
```

```python
# sum of stores count
stores_count = stores_alive.groupby(["行政區","品牌"]).size().unstack(level=-1)
stores_count["超商總數"] = stores_alive.groupby(["行政區"]).size()
stores_count.loc["總計",:] = stores_count.sum()

# and percentage
stores_count = stores_count.fillna(0)
stores_pct = stores_count[['7-11', 'OK', '全家', '萊爾富']].div(stores_count['超商總數'], axis=0)

# calculate correlation about number of store brands
corr_count = population[["人口數(人)","人口密度(人/平方公里)"]].apply(lambda row: pandas.to_numeric(row.str.replace(",","")))
corr_count = pandas.concat([corr_count, stores_count, gender["人口性比例(女=100)"]], axis=1, sort=False)
display(corr_count)
display(corr_count.corr())

# calculate correlation about percentage of store brands
temp = elder[["老年人口(65歲以上)", '人口數(人)']].apply(lambda row: pandas.to_numeric(row.str.replace(",","")))
elder_pct = temp["老年人口(65歲以上)"].div(temp['人口數(人)'], axis=0).rename("老人比例")
corr_pct = pandas.concat([stores_pct, elder_pct, gender["人口性比例(女=100)"], corr_count["人口密度(人/平方公里)"]], axis=1, sort=False)
display(corr_pct)
display(corr_pct.corr())

# concat 4 dataframe
#populatio_data = gender.merge(elder, on=['人口數(人)'], left_index=True).merge(population, on=['人口數(人)'], left_index=True)
#populatio_data["超商數"] = stores_count
#display(populatio_data)

# cast data type
#populatio_data['人口數(人)', '男(人)', '女(人)', '幼年人口(0-14歲)', '工作年齡人口(15-64歲)', '老年人口(65歲以上)', '土地面積(平方公里)', '人口密度(人/平方公里)', '戶數(戶)']
#display(populatio_data.dtypes)

```
